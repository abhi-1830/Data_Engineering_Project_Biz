# EXTERNAL LOCATION CREATION ON BRONZE FOLDER

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS ecom_bronze_loc
URL 'abfss://ecom@ecomstore.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL ecom_cred);
```

```python
%sql
show external locations;
    
describe external location ecom_bronze_loc;
```
